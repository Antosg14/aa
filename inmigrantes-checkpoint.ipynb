```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import chardet

# Lee el contenido del archivo en modo binario (b)
with open('darien.csv', 'rb') as f:
    rawdata = f.read()

# Usa chardet para detectar la codificación
result = chardet.detect(rawdata)
encoding = result['encoding']

print(f"La codificación detectada es: {encoding}")
```

```python


# Lee el archivo CSV con la codificación detectada
df = pd.read_csv('darien.csv', delimiter=';', encoding=encoding)

# Ahora deberías poder trabajar con el DataFrame df sin errores de codificación
print(df)
```

```python
total_column = df.pop('Total')

df.insert(len(df.columns), 'Total', total_column)
```

```python
df = df[df['Pais'] != 'Total']
```

```python
# Eliminar comas de la columna 'Total'
df['Total'] = df['Total'].str.replace(',', '')

# Convertir la columna 'Total' a tipo de dato entero
df['Total'] = df['Total'].astype(int)

# Eliminar comas de los valores en las columnas de meses
columnas_meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
df[columnas_meses] = df[columnas_meses].replace({',': ''}, regex=True)

# Convertir las columnas de meses a tipo de dato entero
df[columnas_meses] = df[columnas_meses].astype(int)
```

### Visualización de datos

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Pais'], df['Total'])
plt.xlabel('País')
plt.ylabel('Total de Inmigrantes')
plt.title('Distribución de Inmigrantes por País')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
for pais in df['Pais']:
    plt.plot(meses, df[df['Pais'] == pais][meses].values[0], label=pais)

plt.xlabel('Mes')
plt.ylabel('Total de Inmigrantes')
plt.title('Tendencia de Inmigrantes por Mes')
plt.legend()
plt.show()
```

```python
for mes in meses:
    pais_max = df[df[mes] == df[mes].max()]['Pais'].values[0]
    print(f"En {mes} el país con más inmigrantes es {pais_max} con {df[mes].max()} inmigrantes.")
```

```python
total_inmigrantes = df['Total'].sum()
print(f"El número total de inmigrantes en todos los meses es: {total_inmigrantes}")
```

```python
plt.figure(figsize=(12, 8))
plt.bar(df['Pais'], df['Ene'], label='Ene', color='skyblue')
plt.bar(df['Pais'], df['Feb'], label='Feb', color='orange', bottom=df['Ene'])
plt.bar(df['Pais'], df['Mar'], label='Mar', color='lightgreen', bottom=df['Ene'] + df['Feb'])
plt.bar(df['Pais'], df['Abr'], label='Abr', color='pink', bottom=df['Ene'] + df['Feb'] + df['Mar'])
plt.bar(df['Pais'], df['May'], label='May', color='purple', bottom=df['Ene'] + df['Feb'] + df['Mar'] + df['Abr'])
plt.bar(df['Pais'], df['Jun'], label='Jun', color='brown', bottom=df['Ene'] + df['Feb'] + df['Mar'] + df['Abr'] + df['May'])

plt.xlabel('País')
plt.ylabel('Total de Inmigrantes')
plt.title('Distribución de Inmigrantes por País y Mes')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Ordenar el DataFrame por la columna 'Total' para obtener los países con más inmigrantes
df_sorted = df.sort_values(by='Total', ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(df_sorted['Pais'][:10], df_sorted['Total'][:10])
plt.xlabel('País')
plt.ylabel('Total de Inmigrantes')
plt.title('Top 10 Países con más Inmigrantes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df_sorted['Total'][:10], labels=df_sorted['Pais'][:10], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Proporción de Inmigrantes por País (Top 10)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for pais in df_sorted['Pais'][:5]:
    plt.plot(df.columns[1:-1], df[df['Pais'] == pais].values[0][1:-1], label=pais)
plt.xlabel('Mes')
plt.ylabel('Total de Inmigrantes')
plt.title('Tendencia de Inmigrantes por País (Top 5)')
plt.legend()
plt.show()
```

```python
# Obtener los nombres de los 5 primeros países con más inmigrantes
top_5_paises = df_sorted['Pais'][:5]

# Filtrar los datos para los 5 primeros países con más inmigrantes
df_top_5 = df[df['Pais'].isin(top_5_paises)]

# Seleccionar solo las columnas de meses para el Heatmap
columnas_meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
df_heatmap = df_top_5[['Pais'] + columnas_meses]

# Configurar el índice del DataFrame para que sea el país
df_heatmap.set_index('Pais', inplace=True)

# Convertir las columnas de meses a tipo de dato entero usando loc
df_heatmap.loc[:, columnas_meses] = df_heatmap.loc[:, columnas_meses].astype(int)

# Crear el heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Heatmap de Inmigrantes por País y Mes (Top 5)')
plt.xlabel('Mes')
plt.ylabel('País')
plt.show()
```
